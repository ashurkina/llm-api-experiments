```python
from dotenv import load_dotenv
import os

from openai import OpenAI

client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)
```

```python
from pydantic import BaseModel


class UserStory(BaseModel):
    story: str


class UserStories(BaseModel):
    user_stories: list[UserStory]
```

```python
review = """
Stop sending stupid notifications about the weather and other nonsense.
I don't mind getting notifications when my car needs a service,
but I don't care about irrelevant things.

I don't want to stop all notifications.
I only want notifications related to my vehicle.
"""
```

```python
instructions = """
You are an experienced Product Manager.

Your task is to convert customer reviews into product user stories.

Rules:
- Identify every distinct customer need.
- Rewrite each need as an independent user story.
- Use the format:
  As a <user>, I want <goal>, so that <benefit>.
- Do not merge multiple needs into one story.
- Rewrite complaints as product requirements.
- If the review contains no actionable requirement, return an empty list.
"""
```

```python
response = client.responses.parse(
    model="gpt-5.5",
    instructions=instructions,
    input=review,
    text_format=UserStories,
)

stories = response.output_parsed
```

```python
for story in stories.user_stories:
    print(story.story)
```

```python
print(response.usage.input_tokens)
print(response.usage.output_tokens)
```
