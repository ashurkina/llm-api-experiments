```python
from google_play_scraper import reviews
from google_play_scraper import Sort

import pandas as pd
```

```python
APP_ID = "com.theaa.android.theaa"
```

```python
result, continuation_token = reviews(
    APP_ID,
    lang="en",
    country="gb",
    sort=Sort.NEWEST,
    count=2000
)
```

```python
result[0]
```

```python
df = pd.DataFrame(result)
```

```python
df.columns
```

```python
df = df[
    [
        "reviewId",
        "content",
        "score",
        "at",
        "thumbsUpCount",
        "appVersion"
    ]
]
```

```python
df.head()
```

```python
df.to_csv(
    "aa_reviews.csv",
    index=False
)
```
